```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
mod=tf.keras.models.load_model('tomato.h5', compile=False)
```

```python
import gradio as gr
```

```python
labels=['Bacterial Spot',
        'Early Blight',
         'healthy',
        'Late Blight',
        'Leaf Mold',
        'Septoria Leaf Spot',
        'Two spotted spider',
        'Target Spot',
        'mosiac virus',
        'Yeallow Leaf Curl Virus'
]
```

```python
def classify_im(inp):
  inp = inp.reshape((-1, 256, 256, 3))
  prediction=mod.predict(inp).flatten()
  m=np.argmax(prediction)
  p=np.sum(prediction)
  max=np.max(prediction)
  perc=(max/p)*100
  pred=labels[m]
  # return f"{Disease: {pred} and Accuracy: {perc}}"
  return pred, perc
```

```python
image1_tensor = gr.Image(width=256, height=256)
label = gr.Label(num_top_classes=1)
# process_image(image1)

iface=gr.Interface(fn=classify_im, inputs=image1_tensor, outputs=['text', 'text'], title="PhytoSense- Tomato", live=True)

iface.launch(debug='True', share="True")
image=gr.Image(shape=(256,256))
```
